# seaborn & pandas

## statistical summary

```python
df = pd.read_csv('https://raw.githubusercontent.com/amongnikol/belajaraDataScienceJCOp/refs/heads/main/course1-pengantarMachineLearning/data/cereal.csv', index_col='name')
df.drop(['mfr', 'type', 'vitamins'], axis=1, inplace=True)
df.head()
```

### missing value

```python
plt.figure(figsize=(15, 15))
sns.heatmap(df.isna(), cmap='Blues')
```

### correlation heatmap

```python
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), cmap='bwr', cbar=False, vmin=-1, vmax=1, annot=True, square=True)
```

### pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'])
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'], kind='reg')
```

```python
sns.jointplot(x= 'potass', y= 'fiber', data=df, marginal_kws={'bins':25}, kind='reg')
```
