```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

# inisialisasi figure dan axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.set(xlim=(0, 10), ylim=(-2, 2), title='sin dan cos', xlabel='x', ylabel='y')
```

## shortcut

```python
# adakalanya kita membutuhkan variabel fig
plt.figure(figsize=(6,6))
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'b')
plt.title('sin dan cos')
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y1, linewidth=0.5, label='sin')
plt.plot(x, y2, markersize=3, label='cos')
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# memaksa skala menjadi sama
plt.axis('equal')
```

# scatter plot

```python
df = pd.read_csv('https://raw.githubusercontent.com/amongnikol/belajaraDataScienceJCOp/refs/heads/main/course1-pengantarMachineLearning/data/circle_data.csv', delimiter=';')
df.head()
```

```python
x = df.x
y = df.y
rad = df.radius

plt.scatter(x, y)
plt.axis('equal')
```

```python
plt.scatter(x, y, c='r', s=10)
plt.axis('equal')
```

```python
plt.scatter(x, y, c=rad, s=rad*90, cmap='seismic')
plt.axis('equal')
```

# subplot

## MATLAB style

```python
plt.figure(figsize=(15, 10))

plt.subplot(231)
plt.hist(numbers, bins=80, density=1)

plt.subplot(233)
plt.hist(numbers, bins=80, density=1)

plt.subplot(235)
plt.scatter(x, y, c='r', s=10)
plt.axis('equal')

plt.subplot(236)
plt.hist(numbers, bins=80, density=1)
```

## Matplotlib Style

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))
ax[0, 0].hist(numbers, bins=80, density=1)
ax[0, 1].axis('off')
ax[1, 0].axis('off')
ax[0, 2].hist(numbers, bins=80, density=1)
ax[1, 1].scatter(x, y, c='r', s=10)
ax[1, 1].axis('equal')
ax[1, 2].hist(numbers, bins=80, density=1)
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10), sharex=True, sharey=True)
ax[0, 0].hist(numbers, bins=80, density=1)
ax[0, 2].hist(numbers, bins=80, density=1)
ax[1, 1].scatter(x, y, c='r', s=10)
ax[1, 2].hist(numbers, bins=80, density=1)
```
